## RFM Analysis

Explore data untuk mengumpulkan kolom order_purchase_timestamp untuk recency, order_item_id untuk frequency, dan payment_value untuk monetary

```python
orderitem_df.head()
```

```python
orderpayment_df.head()
```

```python
orders_df
```

Menggabungkan ketiga tabel

```python
RFM_1_df= pd.merge(
    left=orderpayment_df,
    right=orderitem_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

RFM_1_df.head()
```

```python
RFM_2_df= pd.merge(
    left=RFM_1_df,
    right=orders_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

RFM_2_df.head()
```

Membuat Tabel RFM

```python
RFM_df=RFM_2_df.groupby(by="customer_id",as_index=False).agg({
    "order_purchase_timestamp": "max", #mengambil tanggal order terakhir
    "order_item_id": "nunique",
    "payment_value": "sum"
})
RFM_df.columns = ["customer_id", "max_order_timestamp" , "frequency", "monetary"]
RFM_df.head()
     
```

Menghitung terakhir pelanggan melakukan transaksi

```python
RFM_df["max_order_timestamp"] = RFM_df["max_order_timestamp"].dt.date
recent_date = RFM_2_df["order_purchase_timestamp"].dt.date.max()
RFM_df["recency"] = RFM_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
RFM_df.head()
```

Dapat dilihat dari tabel yang dihasilkan bahwa setiap customer memiliki 3 nilai kriteria yaitu:
1. frequency : dihitung dengan mengambil jumlah unik dari order_item_id untuk setiap customer_id. Kolom order_item_id merepresentasikan jumlah item yang dibeli dalam satu pesanan. Oleh karena itu, nunique digunakan untuk menghitung berapa banyak item yang dibeli oleh setiap pelanggan (customer_id), yang mencerminkan frekuensi pembelian oleh pelanggan.
2. monetary :dihitung dengan menjumlahkan total nilai pembayaran (payment_value) untuk setiap customer_id. Kolom payment_value mewakili jumlah total uang yang dihabiskan oleh setiap pelanggan, yang merupakan nilai transaksi atau pendapatan yang dihasilkan.
3. recency: dihitung untuk setiap pelanggan (customer_id) dengan mengambil selisih antara recent_date (tanggal terbaru) dan tanggal pesanan terakhir (max_order_timestamp) untuk setiap pelanggan. Hasilnya adalah jumlah hari sejak pesanan terakhir pelanggan.

Mengurutkan customer berdasarkan recency, frequency, & monetary score

```python
RFM_df['r_rank'] = RFM_df['recency'].rank(ascending=False)
RFM_df['f_rank'] = RFM_df['frequency'].rank(ascending=True)
RFM_df['m_rank'] = RFM_df['monetary'].rank(ascending=True)

RFM_df.head()
```

```python
RFM_df['r_rank_norm'] = (RFM_df['r_rank']/RFM_df['r_rank'].max())*100
RFM_df['f_rank_norm'] = (RFM_df['f_rank']/RFM_df['f_rank'].max())*100
RFM_df['m_rank_norm'] = (RFM_df['m_rank']/RFM_df['m_rank'].max())*100
 
RFM_df.drop(columns=['r_rank', 'f_rank', 'm_rank'], inplace=True)
 
RFM_df.head()
     
     
```

```python
RFM_df['RFM_score'] = 0.15*RFM_df['r_rank_norm']+0.28 * \
    RFM_df['f_rank_norm']+0.57*RFM_df['m_rank_norm']
RFM_df['RFM_score'] *= 0.05
RFM_df = RFM_df.round(2)
RFM_df[['customer_id', 'RFM_score']].head(7)
```

Penghitungan RFM Score:

RFM_df['r_rank_norm']: Mengambil peringkat ternormalisasi (0-100) dari kolom 'r_rank' yang mewakili peringkat recency.

RFM_df['f_rank_norm']: Mengambil peringkat ternormalisasi (0-100) dari kolom 'f_rank' yang mewakili peringkat frequency.

RFM_df['m_rank_norm']: Mengambil peringkat ternormalisasi (0-100) dari kolom 'm_rank' yang mewakili peringkat monetary.

Pemberian Bobot:

Masing-masing peringkat ternormalisasi ('r_rank_norm', 'f_rank_norm', 'm_rank_norm') diberi bobot sesuai dengan proporsi yang ditentukan: 15% untuk recency, 28% untuk frequency, dan 57% untuk monetary.

Menggabungkan peringkat terbobot untuk mendapatkan RFM Score dengan rumus:
